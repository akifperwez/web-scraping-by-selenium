```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url="https://archive.ics.uci.edu/"
driver.get(url)
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
insert1 = driver.find_element_by_xpath("//b/a[@href='datasets.php']")
insert1.click()
```

```python
urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//p[@class='normal']/b/a")]
for url in urls[:20]:
    time.sleep(5)
    try:         #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//span[@class='heading']/b").text
        des=raw_description.replace("Data Set","")
        Dataset_name.append(des)   #appending dataset name in dataset name list
    except NoSuchElementException :
        Dataset_name.append(" ") #appending the no dataset name if no name is given
        
    try:    #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[1]/td[2]/p[@class='normal']").text      
        Data_type.append(raw_description)#appending datatype in datatype list
    except NoSuchElementException :
        Data_type.append(" ")#appending no dataset type if no type is provided

    try:    #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[3]/td[2]/p[@class='normal']").text      
        Task.append(raw_description)#appending task in task list
    except NoSuchElementException :
        Task.append(" ")#appending no task if no task is given

    try:   #exception handling for nosucheleme
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[2]/td[2]/p[@class='normal']").text      
        Attribute_type.append(raw_description)#appending attribute type in attribute type list
    except NoSuchElementException :
        Attribute_type.append(" ")#appending no attribute type if no type is given
    try:   #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[1]/td[4]/p[@class='normal']").text      
        No_of_instances.append(raw_description)#appending no of instances
    except NoSuchElementException :
        No_of_instances.append(" ")#appending no No of instanced is no instances is given

    try:   #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[2]/td[4]/p[@class='normal']").text      
        No_of_attribute.append(raw_description)#appending  no of attribute
    except NoSuchElementException :
        No_of_attribute.append(" ")#appending no No of attribute if data is not give

    try:   #exception handling for nosuchelementexception
        
        driver.get(url)
        raw_description=driver.find_element_by_xpath("//tbody/tr[2]/td[6]/p[@class='normal']").text 
        des=raw_description[0:4]
        Year.append(raw_description)#appending year
    except NoSuchElementException :
        Year.append(" ")#appending no year if no year is given
```

```python
ML=pd.DataFrame({"Dataset Name":Dataset_name,"Data Type":Data_type,"Task":Task,"Attribute Type":Attribute_type,"No of instances":No_of_instances,
                "No of attribute":No_of_attribute , "Year" : Year})
ML
```
